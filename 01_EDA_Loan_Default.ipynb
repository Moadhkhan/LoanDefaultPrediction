```python
import pandas as pd

# Load CSV file
df = pd.read_csv("Loan_default.csv")

# Display first 5 rows
df.head()
```

```python
# Basic structure
df.info()

# Basic statistics
df.describe(include='all')

# Check for missing values
missing = df.isnull().sum()
missing[missing > 0]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check class distribution
sns.countplot(x='Default', data=df)
plt.title('Loan Default Distribution')
plt.xlabel('Default (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()

# Print class percentages
default_counts = df['Default'].value_counts(normalize=True) * 100
print("Class distribution (%):")
print(default_counts)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = ['Age', 'Income', 'LoanAmount', 'CreditScore', 'MonthsEmployed',
            'InterestRate', 'LoanTerm', 'DTIRatio', 'NumCreditLines']

# Plot histograms with KDE
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cat_cols = ['Education', 'EmploymentType', 'MaritalStatus', 
            'HasMortgage', 'HasDependents', 'LoanPurpose', 'HasCoSigner']

# Plot bar plots for each categorical column
for col in cat_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=col, data=df, order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = ['Age', 'Income', 'LoanAmount', 'CreditScore', 'MonthsEmployed',
            'InterestRate', 'LoanTerm', 'DTIRatio', 'NumCreditLines']

# Boxplots to compare numerical features by Default status
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='Default', y=col, data=df)
    plt.title(f'{col} by Loan Default')
    plt.xlabel('Default')
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```

```python
cat_cols = ['Education', 'EmploymentType', 'MaritalStatus', 
            'HasMortgage', 'HasDependents', 'LoanPurpose', 'HasCoSigner']

# Count plots with hue = Default
for col in cat_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=col, hue='Default', data=df)
    plt.title(f'{col} vs Loan Default')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numerical columns + Default
corr_cols = ['Age', 'Income', 'LoanAmount', 'CreditScore', 'MonthsEmployed',
             'InterestRate', 'LoanTerm', 'DTIRatio', 'NumCreditLines', 'Default']

# Compute correlation matrix
corr_matrix = df[corr_cols].corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features')
plt.tight_layout()
plt.show()
```
